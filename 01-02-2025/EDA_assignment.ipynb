```python
#Cleaning JobDescription and JobRequirement
data['JobDescription'].value_counts()
import re


skills = ['python', 'java', 'sql', 'react', 'node.js', 'c++', 'javascript', 'html', 'css']


# Function to clean the text
def clean_tokenize_lowercase_extract_skills(text):
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove special characters (keeping only letters, digits, and spaces)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # Remove excessive whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    # Tokenize by splitting the text on spaces
    tokens = text.split()
    # Convert tokens to lowercase
    tokens = [token.lower() for token in tokens]
    
    # Extract key skills from the text
    extracted_skills = [skill for skill in skills if skill in tokens]
    
    return tokens, extracted_skills

# Apply the function to both columns
data[['Cleaned_Description', 'Extracted_Skills']] = data['JobDescription'].apply(clean_tokenize_lowercase_extract_skills).apply(pd.Series)
data[['Cleaned_Requirement', 'Extracted_Skills_Requirement']] = data['JobRequirment'].apply(clean_tokenize_lowercase_extract_skills).apply(pd.Series)
```

```python
'''
 Splitting & Extracting Data
â€¢ Split Location into City & Country.
â€¢ Extract numeric salary range from Salary (e.g., "50K-60K USD" â†’
min_salary=50000, max_salary=60000).
'''


import pandas as pd
import re


data = {
    'jobpost': ['AMERIA Investment Consulting Company\r\nJOB TI...', 'International Research & Exchanges Board (IREX)...'],
    'Location': ['New York, USA', 'London, UK'],  # Assuming this column exists
    'Salary': ['50K-60K USD', '40K-50K GBP']  # Assuming this column exists
}

# Create DataFrame
df = pd.DataFrame(data)

# 1. **Splitting `Location` into `City` and `Country`**
df[['City', 'Country']] = df['Location'].str.split(',', expand=True)

# 2. **Extracting numeric salary range from `Salary`**
def extract_salary_range(salary_str):
    # Regex to extract the salary range
    match = re.search(r'(\d+)(K)-(\d+)(K)', salary_str)
    if match:
        min_salary = int(match.group(1)) * 1000  # Convert 'K' to actual number (e.g., 50K -> 50000)
        max_salary = int(match.group(3)) * 1000  # Convert 'K' to actual number (e.g., 60K -> 60000)
        return min_salary, max_salary
    return None, None

# Apply the function to extract min and max salary
df[['min_salary', 'max_salary']] = df['Salary'].apply(lambda x: pd.Series(extract_salary_range(x)))

# Display the updated DataFrame
print(df)
```

```python
'''
Filtering & Categorization
â€¢ Identify IT-related jobs from the IT column.
â€¢ Categorize Eligibility and Audience into structured Education Level /
Experience Ranges.
'''

import pandas as pd

# Sample data (replace this with your actual dataset)
data = {
    'jobpost': ['Software Developer', 'Marketing Manager'],
    'IT': [True, False],
    'Eligibility': ['Bachelor', 'Master'],
    'Audience': ['Fresh Graduate', 'Mid-level']
}

# Create DataFrame
df = pd.DataFrame(data)

# 1. **Identify IT-related jobs from the `IT` column**
# Filter IT-related jobs
it_jobs = df[df['IT'] == True]

# Display IT-related jobs
print("IT-related jobs:")
print(it_jobs)

# 2. **Categorize Eligibility and Audience into Education Level / Experience Ranges**
def categorize_eligibility(eligibility):
    # Define education level categories based on eligibility
    if 'Bachelor' in eligibility:
        return 'Bachelor'
    elif 'Master' in eligibility:
        return 'Master'
    elif 'PhD' in eligibility:
        return 'PhD'
    return 'Unknown'

def categorize_audience(audience):
    # Define experience level categories based on audience
    if 'Fresh Graduate' in audience:
        return '0-2 years'
    elif 'Mid-level' in audience:
        return '2-5 years'
    elif 'Senior' in audience:
        return '5+ years'
    return 'Unknown'

# Apply the categorization functions
df['Education_Level'] = df['Eligibility'].apply(categorize_eligibility)
df['Experience_Range'] = df['Audience'].apply(categorize_audience)

# Display the updated DataFrame
print("\nCategorized Data:")
print(df)
```
