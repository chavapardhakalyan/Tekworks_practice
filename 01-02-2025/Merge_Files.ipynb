```python
import pandas as pd
from sqlalchemy import create_engine

username = 'root'
password = '1234'
host = 'localhost'
database = 'cvr'

connection_string1 = f'mysql+mysqlconnector://{username}:{password}@{host}/{database}'

engine1 = create_engine(connection_string1)

colleges = pd.read_sql('SELECT * FROM Colleges', engine1)

print(colleges)
```

```python
recruitments = pd.read_csv("RecruitmentDrives.csv")
```

```python
recruitments
```

```python
candidates = pd.read_excel("Candidates.xlsx")
```

```python
candidates
```

```python
merged_df = pd.merge(colleges, recruitments, on='CollegeID', how='outer')
```

```python
merged_df
```

```python
merged_final = pd.merge(merged_df, candidates, on=['CollegeID','DriveID'], how='outer')
```

```python
merged_final
```

```python
merged_final.dtypes
```

```python
merged_final.info()
```

```python
print(f"Memory Usage : {merged_final.memory_usage(deep=True)}")
```

```python
import numpy as np
merged_final['CollegeID'] = merged_final['CollegeID'].astype(np.int16)
merged_final['DriveID'] = merged_final['DriveID'].astype(np.int32)
merged_final['CandidateID'] = merged_final['CandidateID'].astype(np.int32)
merged_final['TimeToHire'] = merged_final['TimeToHire'].astype(np.int32)
merged_final['CostPerHire'] = merged_final['CostPerHire'].astype(np.int32)
merged_final['Location'] = merged_final['Location'].astype('category')
```

```python
print(f"Memory Usage : {merged_final.memory_usage(deep=True)}")
```

```python
import numpy as np
data = {'order_id': np.arange(1,1001)}
df = pd.DataFrame(data)

print(df['order_id'].dtype)
print(f"Memory Usage : {df.memory_usage(deep=True)}")
df['order_id'] = df['order_id'].astype(np.int16)

print(f"Memory Usage : {df.memory_usage(deep=True)}")
```

```python
data = {'sales_amount': np.random.random(1000000)}
df = pd.DataFrame(data)

print(df['sales_amount'].dtype)
df['sales_amount'] = df['sales_amount'].astype(np.float32)

print(f"Memory Usage: {df.memory_usage(deep=True)}")
```
